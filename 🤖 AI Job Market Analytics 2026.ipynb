```python
import pandas as pd

df = pd.read_csv("C:/Users/DELL/Downloads/AI Job Market Dataset.csv")

print(df.shape)

print(df.head())

print(df.info())

print(df.describe())
```

Which AI job roles offer the highest average salaries?

```python
# ==========================================
# TOP PAYING AI ROLES ANALYSIS
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt

# Average salary for each job role
top_roles = (
    df.groupby("job_title")["salary"]
    .mean()
    .sort_values(ascending=False)
)

# Print exact values
print("Average Salary by Job Role")
print(top_roles)

# Create visualization
plt.figure(figsize=(8,5))

top_roles.plot(
    kind="bar"
)

plt.title(
    "Average Salary by AI Job Role"
)

plt.xlabel(
    "Job Role"
)

plt.ylabel(
    "Average Salary ($)"
)

plt.xticks(rotation=20)

plt.tight_layout()

plt.show()

# ==========================================
# KEY FINDINGS
# ==========================================

print("\nKey Findings:")
print("1. AI Engineer is the highest paying role ($139,945)")
print("2. ML Engineer is the second highest paying role ($139,705)")
print("3. Data Analyst has the lowest average salary among listed roles")
print("4. AI & ML Engineering roles pay ~40% more than Data Analyst roles")

print("\nBusiness Recommendation:")
print("Professionals seeking higher salaries should focus on AI Engineering and Machine Learning Engineering skills")
```

How does experience influence salary?

```python
salary_exp = (
    df.groupby("experience_level")["salary"]
    .mean()
    .sort_values(ascending=False)
)

print(salary_exp)

salary_exp.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title(
    "Average Salary by Experience Level"
)

plt.ylabel(
    "Salary ($)"
)

plt.show()
```

Does higher education lead to higher salaries?

```python
salary_edu = (
    df.groupby("education_level")["salary"]
    .mean()
    .sort_values(ascending=False)
)

print(salary_edu)

salary_edu.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title(
    "Average Salary by Education Level"
)

plt.ylabel(
    "Salary ($)"
)

plt.show()
```

Which technical skills contribute most to salary growth?

```python
skills_salary = {
    "Python": 113742 - 113143,
    "SQL": 113330 - 113548,
    "Machine Learning": 120626 - 106021,
    "Deep Learning": 121080 - 105857,
    "Cloud": 118154 - 108501
}

skills_df = pd.DataFrame(
    list(skills_salary.items()),
    columns=["Skill","Salary_Boost"]
)

print(skills_df)

skills_df.plot(
    x="Skill",
    y="Salary_Boost",
    kind="bar",
    figsize=(8,5)
)

plt.title(
    "Salary Impact of AI Skills"
)

plt.ylabel(
    "Salary Increase ($)"
)

plt.show()
```

Do remote jobs offer higher salaries?

```python
remote_salary = (
    df.groupby("remote_type")["salary"]
    .mean()
    .sort_values(ascending=False)
)

print(remote_salary)

remote_salary.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title(
    "Average Salary by Work Mode"
)

plt.ylabel(
    "Salary ($)"
)

plt.show()
```

Do urgently hiring companies pay more?

```python
urgency_salary = (
    df.groupby("hiring_urgency")["salary"]
    .mean()
    .sort_values(ascending=False)
)

print(urgency_salary)

urgency_salary.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title(
    "Average Salary by Hiring Urgency"
)

plt.ylabel(
    "Salary ($)"
)

plt.show()
```
